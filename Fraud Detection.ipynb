```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Time', 'V1', 'V2', 'V3', 'V4']] = scaler.fit_transform(df[['Time', 'V1', 'V2', 'V3', 'V4']])

# Split the data into training and testing sets
X = df.drop('Class', axis=1)
y = df['Class']

# Further split the data into training and testing sets (if needed)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
pred = classifier.predict(X_train)
print(classifier.score(X_train,y_train))
```

```python
from sklearn.metrics import classification_report,mean_absolute_error,mean_squared_error,r2_score

report= classification_report(y_train,pred)
print(report)
mean_abs_error = mean_absolute_error(y_train,pred)
mean_abs_percentage_error = np.mean(np.abs((y_train - pred) // y_train))
mse= mean_squared_error(y_train,pred)
r_squared_error = r2_score(y_train,pred)
print("Mean absolute error : {} \nMean Absolute Percentage error : {}\nMean Squared Error : {}\nR Squared Error  {}".format(mean_abs_error,mean_abs_percentage_error,mse,r_squared_error))
```

```python
#Undersampling

from imblearn.under_sampling import RandomUnderSampler
from collections import Counter


# Define your features (X) and target (y)
X = df.drop('Class', axis=1)
y = df['Class']

# Print the original dataset shape
print('Original dataset shape:', Counter(y))

# Define the undersampler
rus = RandomUnderSampler(random_state=42)

# Resample the dataset
X_res, y_res = rus.fit_resample(X, y)

# Print the resampled dataset shape
print('Resampled dataset shape:', Counter(y_res))
```

```python

# Initialize the scaler
scaler = StandardScaler()

# Scale Time and V columns

df[['Time', 'V1', 'V2', 'V3', 'V4']] = scaler.fit_transform(df[['Time', 'V1', 'V2', 'V3', 'V4']])

# Split the data into training and testing sets
X = df.drop('Class', axis=1)
y = df['Class']

# Further split the data into training and testing sets (if needed)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the model
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Hyperparameter tuning for RandomForestClassifier
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Get the best parameters and retrain the model
best_params = grid_search.best_params_
model = grid_search.best_estimator_
```

```python
#try again, after tuning 
# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
